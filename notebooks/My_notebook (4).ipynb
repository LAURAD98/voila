## 1. Introduction

Research question: 

How high was the case fatality rate associated with covid-19 in Germany compared to other European countries in January 2022?


## 2. Data preparation

```python
import pandas as pd
import altair as alt
import ipywidgets as widgets
```

```python
covid = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

After importing the relevant libraries and loading the dataset I first calculated the case fatality rate and then I selected the rows and columns relevant for my research question.

(case_fatality_rate = new_deaths_smoothed_per_million / new_cases_smoothed_per_million * 100)



```python
covid['case_fatality_rate'] = covid['new_deaths_smoothed_per_million'] / covid['new_cases_smoothed_per_million'] * 100
relevant_columns = ['continent', 'location', 'date', 'new_cases_smoothed_per_million', 'new_deaths_smoothed_per_million', 'case_fatality_rate']
europe = covid.loc[covid.continent == 'Europe', relevant_columns]
covid_europe = europe.loc[covid.date.str.contains('2022-01-')]
covid_europe
```

## 3. Exploring and cleaning data

First, I grouped the data by location and used descriptive statistics to get an overview of the values for each country and to see whether there are typos or missing data.


```python
covid_europe.groupby('location').describe()
```

While I did not detect any typos, I saw that for some countries the data is incomplete (NaN).

I decided to delete all rows with missing data.

```python
ce = covid_europe.dropna()
ce
```

After deleting the rows with missing data, I wanted to see which countries had data for every day of January 2022 which is why I looked at the value counts for the countries.

```python
ce['location'].value_counts()
```

While most countries have values for all 31 days of January 2022, the UK, Monaco and Ireland did not which is why I decided to delete the rows for these countries. 
(I only want to include countries with complete data.)

```python
ce = ce[ce['location'].str.contains('United Kingdom|Monaco|Ireland') == False]
ce
```

## 4. Descriptives and visualizations

In order to describe the population, I grouped the data by location (country) and used descriptive statistics.

```python
ce.groupby('location').describe()
```

In the interactive table below, you can select any member of the European Union (except Ireland) to see the new cases, new deaths and case fatality rate for every day of January 2022.


```python
from ipywidgets import interact
from ipywidgets import fixed
```

```python
def select_country_data(location, df):
  return df.loc[df['location'] == location, ]
```

```python
interact(select_country_data, location = ['Germany', 'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France', 'Greece', 'Hungary', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden'], df = fixed(ce))
```

In the interactive chart below, you can select any member of the European Union (except Ireland) to see the case fatality rate (%)for every day of January 2022.


```python
def select_country_data_chart(location, df):
  return df.loc[df['location'] == location, ]
```

```python
def visualize_country_data_chart(location, df):
  visualized_data = select_country_data_chart(location, df)
  
  lineplot = alt.Chart(visualized_data).mark_line().encode(
    x='date:O',
    y='case_fatality_rate:Q',
    color='location:N'
  )

  display(lineplot)
```

```python
interact(visualize_country_data_chart, location = ['Germany', 'Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czechia', 'Denmark', 'Estonia', 'Finland', 'France', 'Greece', 'Hungary', 'Italy', 'Latvia', 'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia', 'Slovenia', 'Spain', 'Sweden'], df = fixed(ce))
```
